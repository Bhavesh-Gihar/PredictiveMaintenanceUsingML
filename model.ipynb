```python
df = pd.read_csv("/home/bhavesh/Documents/AutonomousNavigationAndCollisionAvoidanceBot/bag_files/finalDataSet/finalDataSet.csv")
df.head()
```

```python
X = df.drop(['x', 'y', 'z', 'i', 'j', 'k'], axis=1)
y = df[['x', 'y', 'z', 'i', 'j', 'k']]
```

```python
X.replace([np.inf, -np.inf], np.finfo(np.float32).max, inplace=True)
```

```python
# For xgboost 

mapped_output = []
for index, row in y.iterrows():
    if(row['x']):
        mapped_output.append(0)
    elif(row['k']):
        if(row['k'] == 1):
            mapped_output.append(1)
        else:
            mapped_output.append(2)
    else:
        y.drop(index, inplace=True)
        X.drop(index, inplace=True)

y['output'] = mapped_output
y = y.drop(['x', 'y', 'z', 'i', 'j', 'k'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
# # Create a decision tree classifier
dt_classifier = DecisionTreeClassifier()

# # Define the hyperparameter grid for grid search
# param_grid = {
#     'criterion': ['gini', 'entropy'],
#     'max_depth': [None, 10, 20, 30, 50, 100],
#     'min_samples_split': [2, 5, 10],
#     'min_samples_leaf': [1, 2, 4]
# }

dt_classifier.fit(X_train, y_train)
```

```python
# grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid, cv=5)
# grid_search.fit(X_train, y_train)

# # Get the best parameters and the best model
# best_params = grid_search.best_params_
# best_model = grid_search.best_estimator_
```

```python
rf = RandomForestClassifier(max_depth=100, random_state=0)
rf.fit(X_train, y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
svm = SVC(gamma=0.001)
svm.fit(X_train, y_train)
```

```python
# create model instance
bst = XGBClassifier(n_estimators=1, max_depth=100, learning_rate=1, objective='binary:logistic')
# fit model
bst.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = dt_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy of DT:", accuracy)

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy of KNN:", accuracy)

# Predict on the test set
y_pred = gnb.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy of GNB:", accuracy)

# Predict on the test set
y_pred = svm.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy of SVM:", accuracy)

# Predict on the test set
y_pred = bst.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy of XGB:", accuracy)

# Print the accuracy
print("Accuracy of RF:", accuracy)

# Predict on the test set
y_pred = rf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# y_pred
```
